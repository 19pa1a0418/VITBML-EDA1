# `Description`

This database contains a list of video games with sales greater than 100,000 copies.It was generated by a scrape of "vgchartz.com".fieids include.

**Rank**-    Ranking Of Overall Sales

**Name**-The Games Name


**Platform**-Platform of games relesed (i.e. Ps4,pc,etc.)

**Year**-   Year of the game's relese

**Genre**-Genre of the game(i.e.sports,action etc)

**NA_Sales**-North America sales(in millions)

**EU_Sales**- Europe sales(in millions)

**JP_Sales**- Japan Sales (in millions)

**Other_sales**- Other country sales(in millions)

**Global_sales**- Total worldwide sales

**T**he script to scrape the data is avaliable at (https://github.com/GregorUT/vgchartzScarpe.)

# **Graphical Representation of Data.**

# **Bar Graph.**

```
import matplotlib.pyplot as plt
plt.bar(game['Year'],game['Global_Sales'])
plt.xlabel('Year')
plt.ylabel('Global_Sales(in millions)')
plt.show()
```

#**Box Chart**

```
a=game.head(n=100)
sns.boxplot(x='Year',y='Global_Sales',data=a)
```

# **Scatter Plot**


```
game.plot(x='NA_Sales',y='Global_Sales',kind='scatter')
```

#**Histogram**

```
g3=game.groupby('Year')['Global_Sales'].mean()
g3.plot(kind='hist')
```

# **Joint plot.**

```
sns.jointplot(x='NA_Sales',y='Global_Sales',data=game,kind='scatter')
```

```
g4=game.groupby('Year')['Global_Sales'].mean()
plt.ylabel('globalsales(in millions)')
g4.plot()
```

# **Pie plot.**

```
b=game.head(n=50)
g5=b.groupby('Year')['JP_Sales'].mean()
g5.plot(kind='pie')
```
